Funds by Category

```python
category_counts = df['category'].value_counts()

print(category_counts)

category_counts.plot(kind='bar')

plt.title('Funds by Category')
plt.xlabel('Category')
plt.ylabel('Count')

plt.show()
```

Risk Category Distribution

```python
df['risk_category'].value_counts().plot(kind='bar')

plt.title('Funds by Risk Category')
plt.xlabel('Risk Category')
plt.ylabel('Count')

plt.show()
```

Top 10 Fund Houses Analysis

```python
fund_house_counts = df['fund_house'].value_counts()

print(fund_house_counts)
```

```python
plt.figure(figsize=(10,5))
fund_house_counts.plot(kind='bar')

plt.title('Number of Schemes by Fund House')
plt.xlabel('Fund House')
plt.ylabel('Number of Schemes')

plt.tight_layout()


plt.savefig('../reports/fund_house_distribution.png')

plt.show()
```

Direct vs Regular Plans

```python
plan_counts = df['plan'].value_counts()

print(plan_counts)

plan_counts.plot(kind='bar')

plt.title('Direct vs Regular Plans')
plt.xlabel('Plan')
plt.ylabel('Number of Schemes')
plt.show()
```

Average Expense Ratio by Plan

```python
expense_by_plan = df.groupby('plan')['expense_ratio_pct'].mean()

print(expense_by_plan)

expense_by_plan.plot(kind='bar')

plt.title('Average Expense Ratio by Plan')
plt.xlabel('Plan')
plt.ylabel('Expense Ratio (%)')

plt.show()
```
